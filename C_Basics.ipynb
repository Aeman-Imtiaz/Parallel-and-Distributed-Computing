C structure

```python

%%bash
cat <<EOF > program.c
#include <stdio.h>
#include <string.h>

// Define the structure
struct Student {
    char name[50];
    int age;
    float gpa;
    int id;
};

int main() {
    // Declare a structure variable
    struct Student student1;

    // Access members using dot operator (.)
    strcpy(student1.name, "Aeman ");
    student1.age = 21;
    student1.gpa =3.0;
    student1.id = 001;

    // Print the data
    printf("Student Information:\n");
    printf("Name: %s\n", student1.name);
    printf("Age: %d\n", student1.age);
    printf("GPA: %.2f\n", student1.gpa);
    printf("ID: %d\n", student1.id);

    return 0;
}

EOF
gcc program.c -o program
./program
```

void * in C

A void* is called a generic pointer: It can store address of ANY type
BUT   It CANNOT be directly dereferenced.

 So:
void *ptr;
Means:
“I can point to anything, but I don’t know the type yet.”

 WHY CASTING IS REQUIRED?
*(int*)ptr
Means:
Convert void pointer → int pointer
Then dereference it

```python
%%bash
cat <<EOF > program.c
#include <stdio.h>

int main() {
    int a = 10;
    float b = 5.5;
    char c = 'Z';

    void *ptr;

    ptr = &a;
    printf("Int: %d\n", *(int*)ptr);

    ptr = &b;
    printf("Float: %.1f\n", *(float*)ptr);

    ptr = &c;
    printf("Char: %c\n", *(char*)ptr);

    return 0;
}
EOF

gcc program.c -o program
./program
```

ARRAY USING VOID POINTER

```python
%%bash
cat <<EOF > program.c
#include <stdio.h>

int main() {
    int arr[3] = {10, 20, 30};
    void *ptr = arr;

    for(int i = 0; i < 3; i++) {
        printf("Array[%d]: %d\n", i, *((int*)ptr + i));
    }

    return 0;
}
EOF

gcc program.c -o program
./program
```

GENERIC FUNCTION

```python
%%bash
cat <<EOF > program.c
#include <stdio.h>

void printValue(void *ptr, char type) {

  //   void *ptr: This is a generic pointer. It means: can store address of ANY type but C does NOT know what it points to
 // So you MUST tell C the type before using it.

    if(type == 'i')
        printf("%d\n", *(int*)ptr);

    else if(type == 'f')
        printf("%.2f\n", *(float*)ptr);

    else if(type == 'c')
        printf("%c\n", *(char*)ptr);
}

int main() {
    int x = 10;
    float y = 2.5;
    char z = 'A';

    printValue(&x, 'i');
    printValue(&y, 'f');
    printValue(&z, 'c');

    return 0;
}
EOF

gcc program.c -o program
./program
```

void* return type


```python
%%bash
cat <<EOF > program.c
#include <stdio.h>

// Global variables (live forever)
int global_int = 100;
float global_float = 3.14;
char global_char = 'Z';

// Function that returns a void* to different globals based on input
void* get_global(char type) {
    if (type == 'i')
        return &global_int;      // Returns int* as void*
    else if (type == 'f')
        return &global_float;    // Returns float* as void*
    else if (type == 'c')
        return &global_char;     // Returns char* as void*
    else
        return NULL;
}

int main() {
    // Get integer global
    void* ptr = get_global('i');
    printf("Integer: %d\n", *(int*)ptr);

    // Get float global
    ptr = get_global('f');
    printf("Float: %f\n", *(float*)ptr);

    // Get char global
    ptr = get_global('c');
    printf("Char: %c\n", *(char*)ptr);

    return 0;
}

EOF

gcc program.c -o program
./program
```

Returning int/float/char dynamically (like your current one)

```python
%%bash
cat <<EOF > program.c
#include <stdio.h>

int a = 10;
float b = 3.5;
char c = 'X';

void* getData(char type) {
    if (type == 'i') return &a;
    if (type == 'f') return &b;
    if (type == 'c') return &c;
    return NULL;
}

int main() {
    void* p;

    p = getData('i');
    printf("a= %d\n", *(int*)p);

    p = getData('f');
    printf("b= %.2f\n", *(float*)p);

    p = getData('c');
    printf("c= %c\n", *(char*)p);

    return 0;
}
EOF

gcc program.c -o program
./program
```
