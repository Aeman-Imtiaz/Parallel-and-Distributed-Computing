```python
%%bash
cat <<EOF > code.c
#include <stdio.h>
#include <unistd.h>
#include <sys/types.h>
#include <sys/wait.h>

int main() {
    int pipe1[2];
    int pipe2[2];

    pipe(pipe1);
    pipe(pipe2);

    pid_t pid = fork();

    if (pid == 0) {
        // Child Process

        close(pipe1[0]);
        close(pipe2[1]);

        int num = 5;

        printf("Child: Sending integer %d to Parent\n", num);

        write(pipe1[1], &num, sizeof(num));

        int squared;
        read(pipe2[0], &squared, sizeof(squared));

        printf("Child: Received squared value %d from Parent\n", squared);

        close(pipe1[1]);
        close(pipe2[0]);
    }
    else {
        // Parent Process

        close(pipe1[1]);
        close(pipe2[0]);

        int num;

        read(pipe1[0], &num, sizeof(num));

        printf("Parent: Received integer %d from Child\n", num);

        int squared = num * num;

        printf("Parent: Calculated square = %d\n", squared);

        printf("Parent: Sending squared value %d to Child\n", squared);

        write(pipe2[1], &squared, sizeof(squared));

        wait(NULL);

        close(pipe1[0]);
        close(pipe2[1]);
    }

    return 0;
}
EOF

gcc code.c -o code
./code
```
