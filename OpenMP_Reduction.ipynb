**OpenMP Reduction**

```python
%%bash

cat <<EOF > openmp_sum.c
#include <stdio.h>
#include <omp.h>

#define SIZE 100000

int main() {

    int arr[SIZE];
    long long sum = 0;

    for (int i = 0; i < SIZE; i++) {
        arr[i] = 1;
    }

    int threads_used = 0;

    double start_time = omp_get_wtime();

    #pragma omp parallel
    {
        #pragma omp single
        {
            threads_used = omp_get_num_threads();
            printf("Number of Threads = %d\n", threads_used);
        }

        #pragma omp for reduction(+:sum)
        for (int i = 0; i < SIZE; i++) {
            sum += arr[i];
        }
    }

    double end_time = omp_get_wtime();

    printf("\nFinal Sum = %lld\n", sum);
    printf("Execution Time = %f seconds\n",
           end_time - start_time);

    return 0;
}
EOF

gcc openmp_sum.c -o openmp_sum -fopenmp
./openmp_sum
```
